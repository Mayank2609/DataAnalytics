```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv(r'C:\Users\LENOVO\Desktop\Data Analytics\CSV\AB_NYC_2019.csv')
```

```python
listings
```

```python
# the below plot the prices are grouped into contiguous intervals called bins
plt.hist(listings['price'])
plt.xlabel('price(in us dollars)')
plt.show()
```

```python
plt.hist(listings['price'],bins = range(0,1100,40))
plt.xlabel('price(in us dollars)')
plt.show()
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('Price')
plt.ylabel('Number OF Reviews')
plt.show()
```

```python
# xlim method is used for to restrict x axis with a range
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('Price')
plt.ylabel('Number OF Reviews')
plt.xlim(0,1100)
plt.show()
```

```python
# s is denote the size of scatter means in in the scatter graph each balls size will be 5
plt.scatter(x=listings['price'],y=listings['number_of_reviews'], s=5)
plt.xlabel('Price')
plt.ylabel('Number OF Reviews')
plt.xlim(0,1100)
plt.show()
```
