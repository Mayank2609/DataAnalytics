```python
import pandas as pd
```

```python
us_babies = pd.read_csv("us_baby_names.csv")
```

```python
us_babies
```

```python
us_babies['Year']
```

```python
us_babies['Year'] == 2014
```

```python
# Denotes want to store only 2014 babies data into us_babies_2014 variable 
# ':' sign denotes that access all the rows and columns from that CSV 
us_babies_2014 = us_babies.loc[us_babies['Year']==2014,:]
```

```python
us_babies_2014
```

```python
# For sorting us babies 2014 data we need to use sort_values function 
#under sort_values we are passing two variables 1st is for which column respect we need to sort second is order 
#ascending = False means high to low and ascending = True means low to high
sort_us_babies_2014 = us_babies_2014.sort_values('Count',ascending=False)
```

```python
sort_us_babies_2014
```

```python
#iloc function is using to get specfic row data
#within iloc we have to pass two varialbe 1st is fist index generallly index stat from 0 and 
#last index which one you want here we have choosen 5
sort_us_babies_2014.iloc[0:5]
```
